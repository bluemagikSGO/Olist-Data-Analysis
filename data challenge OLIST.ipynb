```python
# From the look at the unique values in the geolocation city column, we can see that there is a duplicate 
# of sao paulo which is as a result of misspelling, we have to chose one and that will be sao paulo
# to solve this, we have to define a fuction that will make the change for us
```

```python
def sao(x):
    if x == 'são paulo':
        x = 'sao paulo'
        return x
    else:
        return x
```

```python
geolocation["geolocation_city"] = geolocation["geolocation_city"].apply(sao)
```

```python
## from observation, we can see that the shipping limit date is of the object data type which means its just a string
## i will have to change it from being an object Dtype to a datetime Dtype
```

```python
## After changing the datatype to be a datetime, it feels only right to have a different column for the data and time
```

```python
orderitems['shippin_limit_date'] = orderitems.shipping_limit_date.dt.date
orderitems['shippin_limit_time'] = orderitems.shipping_limit_date.dt.time
```

```python
orderitems.head()
```

```python
## After creating the date and time, i have to drop the initial shipping limit date column
```

```python
orderitems = orderitems.drop('shipping_limit_date', axis = 1)
```
